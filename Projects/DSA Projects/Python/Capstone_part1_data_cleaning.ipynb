### Q5.Based on your historic electricity usage, what would your annual cost of  electricitybe under the "Monthly Flex" contract?

```python
month_unique = ['January','February','March', 'April', 'May','June','July','August','September','October','November','December']
```

```python
monthly_rate = [0.2, 0.19, 0.17, 0.18, 0.22, 0.27, 0.24, 0.19, 0.18, 0.15, 0.14, 0.19]

cost = 0
cnt = 0

for x in month_unique:
    cost += df.loc[df['month'] == x]['kwh'].sum() * monthly_rate[cnt]
    cnt += 1

cost    
```

###  Q6.Based on your historic electricity usage, which of the three contracts would pro- duce the lowest annual cost?

```python
rule = pd.read_excel('data_cleaning.xlsx',sheet_name='Contracts')
```

```python
df_rule=rule.iloc[11:35,[4,5,6,8,9]]
```

```python
df_rule.columns=["month","weight_month","EXP","Time","hours_weight"]
```

```python
df_rule
```

```python
df_rule.Time=df_rule.Time.str[:4].str.strip()
```

```python
data = df.copy()
```

```python
data['time'] = data['time'].astype('str')
```

```python
data['time'] = data['time'] + data['AM/PM']
```

```python
data
```

```python
data_months=data.groupby(data.month)['kwh'].sum()
data_hours=data.groupby([data.time])['kwh'].sum()
```

```python
df_rule = df_rule.rename(columns={"Time":'time'})
```

```python
hourly_flex_plan = pd.merge(data_hours, df_rule,on=["time"])
mothly_flex_plan = pd.merge(data_months, df_rule,on=["month"])
```

```python
hourly_flex_plan["sum_hour"]=hourly_flex_plan.kwh*hourly_flex_plan.hours_weight
```

```python
mothly_flex_plan["sum_month"]=mothly_flex_plan.kwh*mothly_flex_plan.weight_month
```

```python
print((mothly_flex_plan.sum_month.sum(),hourly_flex_plan.sum_hour.sum(),data.kwh.sum()*0.21),"The  Hourly Flex  plan is lowest")
```
